### **Introducción**

El Tour de Francia, una de las competiciones deportivas más icónicas y desafiantes del mundo, es un evento anual de ciclismo de carretera que cautiva a audiencias globales desde hace más de un siglo. Esta carrera épica se extiende a lo largo de tres semanas y recorre las espectaculares carreteras de Francia y sus alrededores, atrayendo a los mejores ciclistas del mundo en busca de la gloria y la consagración en el ciclismo.

Desde su primera edición en 1903, el Tour de Francia ha evolucionado en términos de recorrido, reglamento y participantes, pero su esencia única y su impacto en el mundo del deporte se mantienen inquebrantables. La carrera se compone de múltiples etapas que varían en distancia y dificultad, incluyendo contrarrelojes individuales y etapas de montaña desafiantes que ponen a prueba la resistencia, la estrategia y la determinación de los ciclistas.

En este proyecto, nos sumergiremos en la rica historia del Tour de Francia a través de la descripción y análisis de datos estadísticos. Compararemos el rendimiento de los ciclistas y desentrañaremos los factores que contribuyen a la victoria en esta competición, utilizando técnicas de análisis de datos, visualizaciones y modelos estadísticos. Exploraremos los secretos que hacen del Tour de Francia una experiencia deportiva incomparable.

### **Descripción de los datos**

Contamos con *cuatro conjuntos* de datos en archivos planos con instancias y características registradas a lo largo de *119 años* de historia del Tour de Francia (1903 – 2022). \

*Datos de los Tours realizados*, identificando para *109 registros* su año y fecha en la que se desarrolló el evento, número de etapas y la distancia total de la carrera en kilómetros y millas, además del número de ciclistas que iniciaron la carrera y el número de estos que la finalizaron. \

*Datos de las etapas realizadas*, identificando para *2341 registros* el año de la carrera y la fecha en que tuvo lugar la etapa, el número de la etapa y la ciudad de origen y destino, su distancia en kilómetros y millas, el tipo de etapa y el ganador de la misma. \

*Datos de los finalizadores*, identificando para *9895 registros* el año de la carrera, el ranking de clasificación general, el nombre del corredor, el tiempo para terminar la carrera y el equipo o patrocinador del participante. \

*Datos de los ganadores de las carreras*, identificando para *102 registros* el año de la carrera, la nacionalidad y el nombre del ganador, su equipo o patrocinador, el tiempo invertido para terminar la carrera y el margen de victoria sobre el segundo lugar, número de etapas ganadas y número de etapas en las cuales portó la camiseta amarilla, su velocidad media en kilómetros por hora, sus medidas antropométricas como talla y peso, su fecha de nacimiento y de fallecimiento si es el caso.

**Normalización de datos**

```python
# Damos estructura temporal a los datos faltantes en velocidad media, altura, peso y tiempo para su normalización
tdf_winners['Avg Speed'] = tdf_winners['Avg Speed'].fillna(0)
tdf_winners['Height'] = tdf_winners['Height'].fillna(0)
tdf_winners['Weight'] = tdf_winners['Weight'].fillna(0)
tdf_winners['Time'] = tdf_winners['Time'].fillna('0h 0: 0"')

# Se muestra el DataFrame de los ganadores
tdf_winners
```

```python
# Normalización de datos en las variables de velocidad media, estatura y peso de los ganadores

# Limpieza de datos en velocidad media para su manejo numérico
tdf_winners['Avg Speed'] = tdf_winners['Avg Speed'].str.split('km/h',
                                                              expand = True)[0].astype(float)

# Limpieza de datos en la estatura para su manejo numérico
tdf_winners['Height'] = tdf_winners['Height'].str.split('m',
                                                        expand = True)[0].astype(float)

# Limpieza de datos en el peso para su manejo numérico
tdf_winners['Weight'] = tdf_winners['Weight'].str.split('kg',
                                                        expand = True)[0].astype(float)

# Función para la estructuración de los datos en la variable tiempo para obtener el tiempo en horas y su manejo numérico
def time_rider(tdf_winners):
  """
  Esta función recibe como parámetro el dataframe para la lectura de los datos,
  estructura la variable tiempo para obtener el tiempo de los ganadores en horas.
  Retorna el tiempo de los corredores en horas
  """
  s = str(tdf_winners['Time'])
  s = s.replace('+ ','').replace('h ',':').replace("' ", ':').replace('"','').replace(' ','')
  return ((int(s.split(':')[0]) * 3600 + int(s.split(':')[1]) * 60 + int(s.split(':')[2])) / 3600)

# Se llama la función y se aplica la nueva variable Time-H a nuestro DataFrame
tdf_winners['Time-H'] = tdf_winners.apply(time_rider, axis=1)

# Se muestra el dataframe con la nueva columna Time-H
tdf_winners
```

```python
# Normalización de datos de los finalizadores

# Se define la función para identificar en una variable el país del finalizador
def R_Country(tdf_finishers):
  """
  Esta función recibe como parámetro el dataframe para la lectura de los datos,
  estructura la variable del nombre del corredor para obtener de manera separada
  el país de los finalizadores del Tour.
  Retorna una variable con el País del finalizador.
  """
  # Nueva variable para al separación del nombre del corredor y el país
  s = tdf_finishers['Rider']
  # Se realiza la separación del string
  a = s.find(')')
  return(s[a-3:a])

# Se crea la nueva variable en el DataFrame para visualizar el país del finalizador
# Y se hace uso de la función definida
tdf_finishers['Rider_Country'] = tdf_finishers.apply(R_Country, axis = 1)

# Se muestra el DataFrame de los finalizadores con la nueva variable del país del corredor
tdf_finishers
```

**Imputación de datos perdidos**

*Ganadores*

```python
# Imputación con un modelo predictivo a partir de los datos vecinos
from sklearn.impute import KNNImputer

# Construimos el modelo para la imputación de la estatura
imputer = KNNImputer(n_neighbors = 5, # Número de muestras vecinas que se utilizarán para la imputación
                     weights = "uniform") # Pesos uniformes. Todos los puntos en cada vecindario se ponderan por igual.

# Ajustamos el modelo e imputamos los missing values para la estatura de los ganadores
imputer.fit(tdf_winners[["Height"]]) # Carga la variable para imputar
tdf_winners["Height"] = imputer.transform(tdf_winners[["Height"]]).ravel() # Ajusta y transforma los datos

print("Valores perdidos en la estatura de los ganadores es: " +
      str(tdf_winners['Height'].isnull().sum()))


# Ajustamos el modelo e imputamos los missing values para el peso de los ganadores
imputer.fit(tdf_winners[["Weight"]])
tdf_winners["Weight"] = imputer.transform(tdf_winners[["Weight"]]).ravel()

print("Valores perdidos en el peso de los ganadores es: " +
      str(tdf_winners['Weight'].isnull().sum()))


# Limpiamos la variable tiempo para lograr la estructura de datos para la imputación
tdf_winners['Time-H2'] = tdf_winners['Time-H'].replace(0,'NaN')

# Ajustamos el modelo e imputamos los missing values para el tiempo de los ganadores
imputer.fit(tdf_winners[["Time-H2"]])
tdf_winners["Time-H2"] = imputer.transform(tdf_winners[["Time-H2"]]).ravel()

print("Valores perdidos en el tiempo de los ganadores es: " +
      str(tdf_winners['Time-H2'].isnull().sum()))


# Ajustamos el modelo e imputamos los missing values para el tiempo de los ganadores
imputer.fit(tdf_winners[["Avg Speed"]])
tdf_winners["Avg Speed"] = imputer.transform(tdf_winners[["Avg Speed"]]).ravel()

print("Valores perdidos en la velocidad media de los ganadores es: " +
      str(tdf_winners['Avg Speed'].isnull().sum()))


# Mostrar el DataFrame
tdf_winners
```

*Datos relevantes*

```python
# Cual es el top 10 de paises con mayor número de Finalizadores de etapas

# Se identifica el número de finalizadores por país
finishers_country = tdf_finishers.groupby("Rider_Country")["Rider"].count().reset_index()

# Se identifican los primeros 10 países con mayor número de finalizadores en la totalidad de los Tours
finishers_country = finishers_country.sort_values(by = 'Rider',
                                                  ascending = False).head(10)

# Se muestran los los primeros 10 países con mayor número de finalizadores en la totalidad de los Tours
finishers_country
```

```python
# Top 10 de los equipos con mayor número de gandores y su país de origen

d = {"Year": "Conteo Ganadores"}
tdf_winners.pivot_table('Year',
                        index = ['Team', 'Country'],
                        aggfunc = 'count').sort_values('Year',
                                                       ascending = False).rename(columns = d).head(10)
```

```python
# Top 10 de tours con mayor participación de corredores al inicio del Tour

# Se muestran los primeros 5 Tours con mayor participación, mayor participantes entrantes a la carrera
tdf_tours.sort_values('Starters',
                      ascending = False).head(10)
```

```python
# Top 10 de Tours con mayor porcentaje de abandonos,
# es decir el porcentaje que representa la diferencia entre los entrantes y finalizadores del Tour

# Se calcula el porcentaje de abandono para cada Tour
tdf_tours['%_abandonment'] = ((tdf_tours['Starters'] - tdf_tours['Finishers']) / tdf_tours['Starters']) * 100

# Se muestran los primeros 5 Tour con mayor porcentaje de abandono
tdf_tours.sort_values('%_abandonment',
                      ascending = False).head(10)
```

```python
# Top 10 de Corredores que más han ganado el tour, y de qué país es cada uno

d = {"Year": "Conteo Ganadores"}
tdf_winners.pivot_table('Year',
                        index = ['Rider', 'Country'],
                        aggfunc = 'count').sort_values('Year',
                                                       ascending = False).rename(columns = d).head(10)
```

```python
# Identificación de los países de origen de los corredores con mayor número de victorias en la historia del Tour

# Identificando el número de ganadores por país
ganador_pais = tdf_winners.groupby("Country")["Rider"].count().reset_index()
# Identificando los países con mayor número de victorias
ganador_pais = ganador_pais.sort_values(by = 'Rider',
                                        ascending = False)

# Construcción del gráfico
plt.style.use('classic')
plt.figure(figsize = (10, 6))
sns.barplot(data = ganador_pais,
            x = "Country",
            y = "Rider")
plt.title("Victorias por País")
plt.xlabel("País")
plt.ylabel("Número de Victorias")
plt.xticks(rotation = 45)

# Se muestra el gráfico
plt.show()
```

```python
# Distribución de la estatura de los ganadores

# Construcción del gráfico
plt.figure(figsize = (15, 6))
sns.histplot(tdf_winners["Height"],
             bins = 20,
             kde = True)
plt.title("Distribución de la estatura de los Ganadores")
plt.xlabel("Altura (m)")
plt.ylabel("Frecuencia")

# Se muestra el gráfico
plt.show()
```

```python
# Se genera tabla dinámica con el cálculo de la media de la estatura con los países ganadores del Tour

tdf_winners.pivot_table('Height',
                        index = 'Country').sort_values('Height',
                                                       ascending = False).head(10)
```

```python
# Distribución del peso de los ganadores

# Construcción del gráfico
plt.figure(figsize = (15, 6))
sns.histplot(tdf_winners["Weight"],
             bins = 20,
             kde = True)
plt.title("Distribución del peso de los Ganadores")
plt.xlabel("Peso (kg)")
plt.ylabel("Frecuencia")

# Se muestra el gráfico
plt.show()
```

```python
# Se genera tabla dinámica con el cálculo de la media del peso con los países ganadores del Tour

tdf_winners.pivot_table('Weight',
                        index = 'Country').sort_values('Weight',
                                                       ascending = False).head(10)
```
